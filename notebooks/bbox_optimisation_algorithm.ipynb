```python
import numpy as np
import matplotlib.pyplot as plt

joint_lens = np.load('/mnt/nas_siddharth/code_final/calib/_joint_len.npy', allow_pickle=True)
print("joint_len.npy shape:", joint_lens.shape)

mean_len = joint_lens.mean(axis=0)
std_len  = joint_lens.std(axis=0)
labels = [f'J{i}' for i in range(mean_len.size)]

plt.figure(figsize=(12, 4))
plt.bar(labels, mean_len, yerr=std_len)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Length (mm or px)')
plt.title('Mean Joint Lengths from joint_len.npy')
plt.tight_layout()
plt.show()
```

```python
import h5py

intrinsic_file = '/mnt/nas_siddharth/code_final/calib/cam_intrinsic.h5'

with h5py.File(intrinsic_file, 'r') as f:
    print("camera:", list(f.keys()))
    for cam_name in f:
        cam_group = f[cam_name]
        print(f"\n--- Camera: {cam_name} ---")
        print("datasets:", list(cam_group.keys()))
        for dset in cam_group:
            print(f"Dataset '{dset}' shape: {cam_group[dset].shape}")
            if cam_group[dset].shape == (3,3):
                intr = cam_group[dset][:]
                print("Intrinsic matrix:\n", intr)
                print("fx = %.1f, fy = %.1f" % (intr[0,0], intr[1,1]))
                print("Principal point (cx, cy) = (%.1f, %.1f)" % (intr[0,2], intr[1,2]))
```
